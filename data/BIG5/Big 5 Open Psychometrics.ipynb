```python
def calc_pers_score(df):
    score = []
    for row in df.values:
        score.append(row.sum())
    return score

O_columns = ['O1', 'O2', 'O3', 'O4', 'O5', 'O6', 'O7', 'O8', 'O9', 'O10']
C_columns = ['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9', 'C10']
E_columns = ['E1', 'E2', 'E3', 'E4', 'E5', 'E6', 'E7', 'E8', 'E9', 'E10']
A_columns = ['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7', 'A8', 'A9', 'A10']
N_columns = ['N1', 'N2', 'N3', 'N4', 'N5', 'N6', 'N7', 'N8', 'N9', 'N10']

df['O_score'] = calc_pers_score(df[O_columns])
df['C_score'] = calc_pers_score(df[C_columns])
df['E_score'] = calc_pers_score(df[E_columns])
df['A_score'] = calc_pers_score(df[A_columns])
df['N_score'] = calc_pers_score(df[N_columns])
```

```python
possible_percentiles = list(range(101))
    
def calc_percentile(score, pop_scores):
    score_percentiles = []
    
    for perc in possible_percentiles:
        score_percentiles.append(np.percentile(pop_scores, perc))
        
    calc_perc = 0
    for score_perc, perc in zip(score_percentiles, possible_percentiles):
        if int(score_perc) >= score:
            calc_perc = perc
            break
        elif int(score_perc) > score:
            break
            
    return calc_perc
```

```python
import scipy.stats as stats

df['O_percentile'] = df['O_score'].apply(lambda x: stats.percentileofscore(df['O_score'].sort_values(),x))
df['C_percentile'] = df['C_score'].apply(lambda x: stats.percentileofscore(df['C_score'].sort_values(),x))
df['E_percentile'] = df['E_score'].apply(lambda x: stats.percentileofscore(df['E_score'].sort_values(),x))
df['A_percentile'] = df['A_score'].apply(lambda x: stats.percentileofscore(df['A_score'].sort_values(),x))
df['N_percentile'] = df['N_score'].apply(lambda x: stats.percentileofscore(df['N_score'].sort_values(),x))
df
```
