```python
# Models for Multilabel classification
X = tfidf.fit_transform(df['STATUS'])
y = df[['cOPN', 'cCON', 'cEXT', 'cAGR', 'cNEU']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=32)

print('Model Performance for Multilabel prediction:')

forest = RandomForestClassifier(random_state=32)
multi_target_forest = MultiOutputClassifier(forest, n_jobs=-1)
multi_target_forest.fit(X, y)
score = np.mean(cross_validate(multi_target_forest, X_test, y_test, cv=10)['test_score'])
print('Multi Random Forest Score: ' + str(score))
```

```python
RFR = RandomForestRegressor()
RFR.fit(X_train, y_train)

RFR_score = np.mean(cross_validate(RFR, X_test, y_test, cv=5)['test_score'])
RFR_score
```
